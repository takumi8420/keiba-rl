```python
import json
import pandas as pd
import boto3
from io import StringIO
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from time import sleep
from selenium.webdriver.common.by import By
```

```python
def get_race_table(driver, race_id):
    url = f"https://race.netkeiba.com/race/shutuba.html?race_id={race_id}" 
    driver.get(url)
    sleep(2)
    # テーブルを取得
    table = driver.find_element(By.CLASS_NAME, 'Shutuba_Table')

    # 各行を取得
    rows = table.find_elements(By.CLASS_NAME, 'HorseList')

    # テーブルデータを格納するリスト
    table_data = []

    # 各行の情報を取得
    for row in rows:
        # 各セル(tdタグ)を取得
        cells = row.find_elements(By.TAG_NAME, 'td')
        row_data = [cell.text for cell in cells]
        table_data.append(row_data)

    # データフレームに変換
    df = pd.DataFrame(table_data)

    # データフレームを表示
    df=df[[3,4,5,6,7,9,10]]

    df.rename(columns={3: '馬名', 4: '性齢', 5: '斤量', 6: '騎手', 7: '厩舎', 9: '予想オッズ', 10: '人気'}, inplace=True)
    return df
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")

driver = webdriver.Chrome(options=chrome_options)
get_race_table(driver, 202405030611)
```
